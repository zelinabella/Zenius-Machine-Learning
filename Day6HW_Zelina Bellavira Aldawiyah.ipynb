**1. Create Pandas DataFrame**

```python
import pandas as pd
import numpy as np
```

```python
df1 = pd.DataFrame({
    "name": ["Anna", "Dane", "David", "Kevin", "Joe", "Rose"],
    "weight": [45.2, 50.5, 62.1, 64.0, 59.0, np.nan],
    "height": [155, 160, 162, 170, 167, 159],
    "birth_year": [2000,1997,1996,1987,1998,2003],
    "gender": ["female", "male", "male", "male", "male", "female"],
    "city": ["New York", "Chicago", "Austin", "New York", np.nan, "New York"]
    })
df1
```

**2. Create another Pandas DataFrame and merge it with the previous DataFrame**

```python
#Membuat DataFrame baru "df2" yang berisi data nick_name dan full_name
df2 = pd.DataFrame({
    "nick_name": ["Anna","David","Dane","Rose"],
    "full_name": ["Anna Levis","David John","Dane Sebastian","Rose Kimberly"],
    })
df2
```

```python
#Merge df1 dan df2
df3=df1.merge(df2,how="outer",left_on="name",right_on="nick_name")
df3
```

**3. Drop nick_name column and fill null values in the full_name**

```python
#Menghilangkan kolom nick_name dan mengisi NaN pada full_name dengan nama
df3=df3.drop(columns = "nick_name") 
df3["full_name"]=df3["full_name"].fillna(df3["name"])
df3
```

**4. Create a new ‘bmi’ column**

```python
#Membuat kolom baru pada DataFrame "df3" berisi hasil perhitungan BMI yang telah dirumuskan 
df3["BMI"]=(df3["weight"])/((df3["height"]/100)**2)
df3
```

**5. Create a new ‘bmi_class’ column**

```python
# Membuat fungsi calculation menggunakan loop untuk menentukan klasifikasi BMI
def calculation(BMI):
    if BMI >= 40:
        return 'Obesity Class 3'
    elif 40 > BMI >= 35:
        return 'Obesity Class 2'
    elif 35 > BMI >= 30:
        return 'Obesity Class 1'
    elif 30 > BMI >= 25:
        return 'Overweight'
    elif 25 > BMI >= 18.5:
        return 'Normal Weight'
    elif BMI < 18.5:
        return 'Underweight'
```

```python
#Membuat kolom baru "BMI_Class" pada DataFrame "df3" yang merupakan hasil klasifikasi BMI
df3["BMI_Class"]=df3.apply(lambda x: calculation(x['BMI']),axis=1)
df3
```
